# ID3 Decision Tree

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics,tree
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report,confusion_matrix

data=pd.read_csv('layeggs.csv')

labelencoder=LabelEncoder()

newdata=pd.DataFrame()

X=data[['Animal','Warmblooded','Feathers','Fur','Swims']]
y=data[['Class']]

x=X
for value in x:
    newdata[value]=labelencoder.fit_transform(data[value])

print("Encoded Data:\n")
print(newdata)

clf=DecisionTreeClassifier(criterion="entropy",max_depth=3)
clf=clf.fit(newdata,y)

y_pred=clf.predict(newdata)

cm = metrics.confusion_matrix(y,y_pred)

print("\nAccuracy:",metrics.accuracy_score(y,y_pred))
print("\nConfusion Matrix:\n",metrics.confusion_matrix(y,y_pred))
print("\nPrediction:\n",y_pred)
print("\nClassification Report:\n",metrics.classification_report(y,y_pred))

print("\t\t\t\t\t Decision Tree\n")
tree = tree.plot_tree(clf,feature_names=x.columns,class_names=data["Class"], filled=True)
```

# C4.5 Decision Tree

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics,tree
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report,confusion_matrix

data=pd.read_csv('layeggs.csv')

labelencoder=LabelEncoder()

newdata=pd.DataFrame()

X=data[['Animal','Warmblooded','Feathers','Fur','Swims']]
y=data[['Class']]

x=X
for value in x:
    newdata[value]=labelencoder.fit_transform(data[value])

print("Encoded Data:\n")
print(newdata)

clf=DecisionTreeClassifier(criterion="gini",max_depth=3)
clf=clf.fit(newdata,y)

y_pred=clf.predict(newdata)

print("\nAccuracy:",metrics.accuracy_score(y,y_pred))
print("\nConfusion Matrix:\n",metrics.confusion_matrix(y,y_pred))
print("\nPrediction:\n",y_pred)
print("\nClassification Report:\n",metrics.classification_report(y,y_pred))


print("\t\t\t\t\t Decision Tree\n")
tree = tree.plot_tree(clf,feature_names=x.columns,class_names=data["Class"], filled=True)
```
