<p style="font-size:1.2em"><b> Aim: AND gate using Perceptron Learning
</b></p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('and.csv',header=None)

data=data.values

pairs=data.shape[0]

print("pairs:",pairs)

attributes=data.shape[1]-1

W=2 * np.random.random_sample((attributes)) - 1

bias=np.random.random()

print("Weight:",W)
print(W.shape)
print("Bias:",bias)
```

```python
lr=0.1
iterations=50

for j in range(iterations):
    for i in range(pairs):
        net_input=np.dot(W,data[i,0:attributes]) + bias
        
        a=net_input>=0
        
        error= data[i,attributes] - a
        
        W = W + lr * error * (data[i,0:attributes].T)
        
        bias = bias + lr * error
        
        print("Iteration :", j , "-",i)
        print("Weight:",W)
        print("Bias:",bias)
        print("\n")

print("Final Weight: ", W)
print("Final Bias: ", bias)

prediction = (np.dot(data[:,0:attributes],W)+bias)>=0

print(prediction)
```
